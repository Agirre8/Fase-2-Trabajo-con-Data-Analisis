```python
columnas_drop= ["Series_Title","Star1", "Series_numerico", "Genre_1", "Genre_2","Director","Genre_0"]

data_filt_3 = data_filt_3.drop(columns=columnas_drop)
```

```python
data_filt_3.dtypes
```

```python
# Generar la lista de features y la variable target
target = 'IMDB_Rating'
features = [x for x in data_filt_3.columns if x!=target]

print(target)
print(features)
```

```python
data_filt_3[target]
```

```python
#Teniamos un dato NAN que hemos eliminado
data_filt_3 = data_filt_3.dropna()
```

```python
data_filt_3.isna().sum()
```

```python
# Importar el algoritmo de árboles de decisión
from sklearn.tree import DecisionTreeRegressor

# Asignar el algortimo e indicar la profundidad máxima del árbol (con un número rotandamente grande para sobreajustar)
arbol_importancia = DecisionTreeRegressor(max_depth=len(features)+10, random_state=100)

# Entrenar un árbol con todo el conjunto de datos
arbol_importancia.fit(X=data_filt_3[features], y=data_filt_3[target])
```

```python
y_pred_arbol = arbol_importancia.predict(X=data_filt_3[features])

from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_absolute_percentage_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

# Métricas para evaluar la calidad del modelo
print('Mean Absolute Error:', mean_absolute_error(data_filt_3[target], y_pred_arbol))
print('Mean Absolute Percentage Error:', mean_absolute_percentage_error(data_filt_3[target], y_pred_arbol)*100)
print('Mean Squared Error:', mean_squared_error(data_filt_3[target], y_pred_arbol))
print('Root Mean Squared Error:', np.sqrt(mean_squared_error(data_filt_3[target], y_pred_arbol)))
print('R^2 coefficient of determination:', r2_score(data_filt_3[target], y_pred_arbol))
```

```python
#Te da los porcentajes de importancia, siemrpe va a sumar 1

arbol_importancia.feature_importances_
```

```python
# Creamos un DataFrame con los datos de importancia
importancia = pd.DataFrame(arbol_importancia.feature_importances_, index=features, columns=["Importancia"])

# Ordenamos los datos
importancia.sort_values(by=importancia.columns[0], ascending=False, inplace=True)
importancia
```
